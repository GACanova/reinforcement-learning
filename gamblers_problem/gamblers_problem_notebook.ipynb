# Gambler’s Problem

This notebook applies the value iteration algorithm to Gambler's Problem, a fundamental scenario in reinforcement learning from Sutton and Barto's book. We aim to find an optimal betting strategy in a coin-flip gambling scenario.

## Import Libraries

```python
import matplotlib.pyplot as plt
from IPython.display import Image
from gamblers_problem_class import GamblersProblem
```

## Run Simulation

```python
gamblers_problem = GamblersProblem(p_h=0.4)
value_function, policy = gamblers_problem.value_iteration()
```

## Plot Value Function

```python
plt.title("Value Function", fontsize=15)
plt.xlabel("Capital", fontsize=15)
plt.ylabel("Value Estimates", fontsize=15)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xticks([1, 25, 50, 75, 99])
plt.plot(value_function, lw=2);
```

## Plot Final Policy

```python
plt.stem(policy)
plt.title("Final Policy", fontsize=15)
plt.xlabel("Capital", fontsize=15)
plt.ylabel("Stake", fontsize=15)
plt.xticks([1, 25, 50, 75, 99])
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.stem(policy)
```

## Comparing to the Book 

```python
Image(filename='gamblers_policy.png')
```

The final policy values in this simulation differ from those in Barto's book (figure above). This variation arises because the problem admits multiple optimal solutions, each with more than one viable greedy action. In our code, ties among these actions are broken randomly, leading to different results in each simulation run. The solution depicted in the book is just one among several ideal solutions. The closest approximation I achieved through iterations is illustrated in the figure below.

```python
seed = 697158

gamblers_problem = GamblersProblem(p_h=0.4, seed=seed)
value_function, policy = gamblers_problem.value_iteration()

plt.stem(policy)
plt.title("Final Policy", fontsize=15)
plt.xlabel("Capital", fontsize=15)
plt.ylabel("Stake", fontsize=15)
plt.xticks([1, 25, 50, 75, 99])
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.stem(policy)
```
