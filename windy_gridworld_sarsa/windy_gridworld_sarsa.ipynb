# Windy Gridworld

<span style="font-family: monospace; font-size:18px; color: white;">
In this notebook, we explore Problem 6.5 from "Reinforcement Learning: An Introduction" by Sutton and Barto, known as the Windy Gridworld challenge. Here, an agent must navigate to a goal in a windy grid. We use the Sarsa algorithm to teach the agent how to find its way. Our aim is to see how the agent learns to reach the goal despite the wind. We'll show the agent's progress and how its path improves over time.
</span>

## Import libraries

```python
from sarsa_agent_class import SarsaAgent
from windy_gridworld_class import WindyGridworld
from utils import get_optimal_value_function_and_policy, episodes_per_timestep, plot_optimal_policy_and_state_values, find_optimal_trajectory, plot_trajectory
from config import ACTIONS, KINGS_ACTIONS, KINGS_ACTIONS_STOP, STATES
```

## Create environment

```python
env = WindyGridworld()
```

## 4 Actions (Up, Down, Right, Left)

### Run simulation

```python
%%time
agent = SarsaAgent(env, ACTIONS, alpha=0.5, epsilon=0.1)
state_action_function, history = agent.run_simulation(n_episodes=10000)
```

### Episodes per timestep

```python
episodes_per_timestep(history)
```

### Optimal State Values and Optimal Policy

```python
state_value_function, optimal_policy = get_optimal_value_function_and_policy(state_action_function, STATES, ACTIONS)
plot_optimal_policy_and_state_values(state_value_function, optimal_policy, ACTIONS)
```

### Optimal trajectory

```python
env.reset()
optimal_trajectory = find_optimal_trajectory(env, optimal_policy, ACTIONS)
plot_trajectory(optimal_trajectory)
```
